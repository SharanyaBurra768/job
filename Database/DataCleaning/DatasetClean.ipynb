```python
file_path = r"/Users/pujithajujjarapu/Documents/680JobDatasSet.csv"
df = pd.read_csv(file_path, encoding="utf-8", engine="python", on_bad_lines="skip")
df.head()
```

```python
df.duplicated().sum()  # Check duplicate rows
df.drop_duplicates(inplace=True)  # Remove duplicate rows
df
```

```python
df
```

```python
columns_to_modify = ['max_salary', 'views', 'med_salary', 'min_salary', 'applies', 'remote_allowed', 'normalized_salary', 'closed_time']
df[columns_to_modify] = df[columns_to_modify].fillna(df[columns_to_modify].mean())
df
```

```python
df = df.round({'max_salary': 1, 'med_salary': 1, 'normalized_salary': 1})
df
```

```python
import datetime
# If the column is in datetime format, convert it to nanoseconds (if needed)

if pd.api.types.is_datetime64_any_dtype(df['expiry']):
    df['expiry'] = df['expiry'].astype('int64')  # Convert to integer type (nanoseconds)

# If the column is numeric, divide the nanoseconds by 1,000,000 to convert to milliseconds
df['expiry'] = df['expiry'] / 1_000_000  # Convert to milliseconds

# Now convert to datetime (milliseconds)
df['expiry'] = pd.to_datetime(df['expiry'], unit='ms')

df
```
